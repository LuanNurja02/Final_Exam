**Compute**: 
The "cosine similarity" between the query vector `query1` and an orthogonal basis of the column space of $A$ (use the QR factorization with pivot: `[Q,R,P]=scipy.linalg.qr(A,mode='economic',pivoting=True)`). 

In particular, determine a suitable subspace spanned by $k$ elements rather than the full set of $\text{rank}(A)$ elements by setting the tolerance to $0.92$ and choosing $k$ following these instructions:

* **Extract** the diagonal elements of $R$ and copy them in an auxiliary vector `Rdiag`;
* **Scale** the absolute values of `Rdiag` with respect to its absolute maximum;
* **Compute** $k$ as the number of elements of `Rdiag` that are greater than the chosen tolerance.

```python
print('Rank of the matrix An:', np.linalg.matrix_rank(An))
```

Since A is nonsingular (fullrank), we can apply the QR factorization. In QR factorization the given matrix A can be represented as the product of an orthogonal matrix Q and an upper triangular matrix R

```python
import scipy.linalg as spl

[Q,R,P]=spl.qr(An, mode = 'economic', pivoting=True)
print('Shape of Q: ', Q.shape)
print('Shape of R: ', R.shape)
```

### THIN QR FACTORIZATION 
*(Also called the Economic Factorization)*

Computes only the $Q1$ and $R$ matrices such that:
* The columns of $Q1$ form an **orthonormal basis** for $\mathcal{R}(A)$ (the range or column space of $A$).

```python
cosin_Q = np.dot(Q.T, query1.T)/np.linalg.norm(query1,2)

print('Cosin of theta for Q: \n', cosin_Q)
print('\nShape: ', cosin_Q.shape)
```

```python
Rdiag = np.diag(R)

scaled_Rdiag = np.abs(Rdiag) / np.max(np.abs(Rdiag))

print('Scaled_Rdiag:\n', scaled_Rdiag)

tolerance = 0.92

k = np.sum(scaled_Rdiag > tolerance)

print('\nThe selected k: ', k)
```

```python
Qk = np.copy(Q[:, :k])
# Qk basis for the column space of A
print('Shape of Qk: ', Qk.shape)

Rk = np.copy(R[:k, :])
print('Shape of Rk: ', Rk.shape)

```

```python
cosin_Qk = np.dot(Qk.T, query1.T)/np.linalg.norm(query1,2)
print('Cosin of theta for Qk: \n', cosin_Qk)
```

 We calculate the Singular value decomposition **SVD** of the matrix **A**, select a suitable **k** and then calculate the cosine similarity between the query and the new approximated matrix

```python
# Full_matrices = false -> allows us to not consider the zero singular values
U, S, Vh = spl.svd(An, full_matrices=False)

print('Shape of U: ', U.shape)
print('Shape of S: ', S.shape)
print('Shape of V^T: ', Vh.shape)
```

```python
#Scree plot method
import matplotlib.pyplot as plt

plt.semilogy(S,'o')
plt.title('Scree plot of Singular Values')
plt.xlabel('Index')
plt.ylabel('Singular Value')
plt.show()
```

```python
# Relative error method
tol = 5e-1

# Ratio between sigma_{k+1} and sigma_1 for each k
ratios = S[1:] / S[0]

plt.plot(range(1, len(ratios)+1), ratios, marker='o', linestyle='-', color='b')
plt.axhline(y=tol, color='r', linestyle='--', label='Threshold')

plt.xlabel('k')
plt.ylabel('Ratio')
plt.title('Relative Error Method')
plt.legend()

plt.show()


# Find the value of k where the ratio is less than or equal to tol
k = np.argmax(ratios <= tol) + 1

print('\nThe value of k where ratio <= ', tol, 'is:', k)
```

```python
# Kaiser rule method
# Compute mean variance
sigma_mean = np.mean(S**2)

k_kaiser = np.sum(S**2 > sigma_mean)

print('Mean variance', sigma_mean)
print('Smallest k with Kaiser rule: ', k_kaiser)
```

```python
# Cumulative variance method
cumulative_sum_k = np.cumsum(S**2)
total_sum = np.sum(S**2)

threshold = 0.7

for k in range(1, len(S) + 1):
    Ek = cumulative_sum_k[k - 1] / total_sum
    if Ek >= threshold:
        break

print('Desired cumulative variance: ', threshold)
print('Smallest k: ', k)
print('Value of Ek: ', Ek)
```

Then we consider **k=8** and build the approximated matrix **Ak**

According to the Eckart-Young-Mirsky Theorem, the optimal rank-$k$ approximation of a matrix $A$ is obtained by truncating its Singular Value Decomposition (SVD) to the first $k$ components:$$A \approx A_k = U_k \Sigma_k V_k^T$$

```python
# According to error relative method
k = 8

# Truncate SVD to retain only the top k components
Uk = np.copy(U[:, :k])
print('Shape of Uk: ', Uk.shape)

Sk = np.copy(np.diag(S[:k]))
print('Shape of Sk: ', Sk.shape)

Vhk = np.copy(Vh[:k, :])
print('Shape of V^Tk: ', Vhk.shape)
```

```python
print(np.linalg.norm((A-Ak),2))
print(np.linalg.norm((A),2))
print(np.linalg.norm((A-Ak),2)/np.linalg.norm((A),2))
```

In the context of Latent Semantic Analysis (LSA), we use the first $k$ columns of $U$ as an orthonormal basis to approximate the document vectors.



```python
Dk = np.dot(Sk, Vhk)
Ak = np.dot(Uk, Dk)
print('Shape of Ak: ', Ak.shape)
```

```python
rank_Ak = np.linalg.matrix_rank(Ak)

print("Rank of Ak with", k, "singular values:", rank_Ak)
```

When using a rank-$k$ approximation, the term-document matrix is represented as $A_k = U_k D_k$. To perform query matching, we compute the relationship between a query vector $q$ and the approximated matrix:$$q^T A_k = q^T U_k D_k = (U_k^T q)^T D_k$$

 The similarity between the query and a document $d_j$ is determined by computing the cosine of the angle $\theta_j$ between them:$$\cos \theta_j = \frac{q_k^T d_j}{\|q_k\|_2 \|d_j\|_2}$$

 with  $q_k = U_k^T q$.

```python
print(query1.shape)
```

```python
qk = np.dot(Uk.T, query1.T)
print(qk)
```

```python
lsi_qk = np.zeros(Dk.shape[1])
qk_flat = qk.flatten() # Ensures qk is 1D
qk_norm = np.linalg.norm(qk_flat, 2)

for i in range(Dk.shape[1]):
    # Dot product of two 1D arrays returns a true scalar
    lsi_qk[i] = np.dot(qk_flat, Dk[:, i]) / (np.linalg.norm(Dk[:, i], 2) * qk_norm)
    print(f'For document at column {i+1}, the LSI with query1 is {lsi_qk[i]}')
```

```python
lsi_qk = lsi_qk.reshape(-1, 1)
print(lsi_qk.shape)
```

```python
print('Cosine between An and query1: \n', cosin)
print('\nLSI between Ak and qk: \n', lsi_qk)
```

Analysis of LSI Results vs. Original Document Matching

Document 7 (Index 6) shows a cosine of 0.0 in the original matrix but jumps to 0.783 in the LSI model.

By truncating the SVD, we remove "noise". This places documents that share a topic much closer together in the $k$-dimensional space.

Let's quantify, now, the difference between the cosine obtained with the complete matrix and the LSI. To do this we apply three measures: the standard deviation, the means of the squared error and the cosine. The ideal value for the first two measurements is 0, while for the calculation of cosine the ideal value is 1.


```python
difference = cosin - lsi_qk

std_dev = np.std(difference)

mse = np.mean((difference)**2)

cosine_theta = np.dot(lsi_qk.T, cosin) / (np.linalg.norm(lsi_qk,2) * np.linalg.norm(cosin,2))


print('Error between LSI and cosine similarity computed using the full matriz A, expressed in terms of:\n')
print('Standard deviation: ', std_dev)
print('Mean Squared Error: ', mse)
print('Cosine of the angle between the two vectors: ', cosine_theta)
```

**Comments**

1. **Standard Deviation: 0.235**
The standard deviation is a measure of data dispersion, that is, how much the values of the two vectors deviate from their mean. Since the difference between the two vectors is zero if the vectors coincide, then the closer the standard deviation approaches zero the more similar the vectors are.
 A standard deviation of 0.235 on a metric like cosine similarity (which ranges from -1 to 1) is quite high.



2. **Mean Squared Error (MSE): 0.074**
The MSE measures the average of the squares of the errors.A higher value indicates greater discrepancy between the two vectors, a value of 0.074 is relatively low in absolute terms, suggesting that the LSI reconstruction is "on the right track".

3. **Cosine of the angle (Cosine Similarity between vectors): 0.736**
The cosine of the angle between the two vectors can can take values from -1 to 1. When the cosine is close to 1, the vectors are aligned, when close to 0, the vectors are orthogonal, when close to -1, the vectors are opposite.
A value of 0.736 indicates a moderate to strong positive correlation, but it is not excellent. Maybe the dimension **k** is to low.


Using the singular value decomposition, compute the closest point to the query in the range of **A** and in the range of **Ak**.

This theorem is known as the **Theorem of the closest point**. Given $M$ a subspace of inner product space $V$ and let $b$ be a vector in $V$.The unique vector in $M$ that is closest to $b$ is $P_M b$, the orthogonal projection of $b$ onto $M$ along $M^\perp$.This theorem says that in text mining:$$\| q - q_A \|_2 \leq \| q - x \|_2 \quad \forall x \in R(A)$$In the theorem:$$\| b - P_M b \|_2 \leq \| b - x \|_2 \quad \forall x \in M$$

Using the SVD decomposition $A = U \Sigma V^T$, the matrix $U_1$ consists of orthonormal bases for the range of $A$. The orthogonal projector onto this subspace is defined as:$$P_{R(A)} = U_1 U_1^T \text{}$$

Similarly, the matrix $V_1$ constitutes the orthonormal basis for $R(A^T)$. The orthogonal projector onto the row space is defined as:$$P_{R(A^T)} = V_1 V_1^T \text{}$$

```python
#Compute the projectors
projector_RA = np.dot(U, U.T)
projector_RAk = np.dot(Uk, Uk.T)

#Compute the closest points
closest_point_A = np.dot(projector_RA, query1.T)
closest_point_Ak = np.dot(projector_RAk, query1.T)
```

```python
#Compute the orthogonal distances
orth_dist_A = np.linalg.norm((query1 - closest_point_A),2)

#Compute the orthogonal distances
orth_dist_Ak = np.linalg.norm((query1 - closest_point_Ak),2)

print(orth_dist_A)
print(orth_dist_Ak)
```

The similarity in orthogonal distances suggests that the dimensionality reduction performed by **Ak**
 effectively preserves the main structure of the relationship between the columns of A, resulting in similar projections of the query in both cases.

Do the eigenvalues of such a matrix satisfy the theoretical relation with the singular value of the matrix $A$?The covariance indicates the level to which two variables vary together. The covariance matrix of a matrix $A_{m \times n}$ is defined as $C = \frac{X^T X}{m-1}$, where $X$ is the centered version of $A$, i.e., for each column $x_i$ in $X$: $x_i = a_i - \mu(a_i)$ with $\mu$ being the mean.Replacing $X$ with its SVD into the definition of the covariance matrix, we get:$$C = \frac{X^T X}{m-1} = \frac{(U S V^T)^T U S V^T}{m-1} = \frac{V \Sigma^2 V^T}{m-1}$$

A straightforward way to determine if a symmetric matrix $A$ is positive definite is to verify that all of its eigenvalues ($\lambda$) are strictly positive ($\lambda > 0$).

You can examine these properties using Singular Value Decomposition (SVD). For a symmetric matrix $A$, the singular values are the absolute values of the eigenvalues.If we look at the product $A^TA$, we can derive the relationship as follows:Decomposition: Using SVD, we express $A = U\Sigma V^T$.Substitution:$$A^TA = (V\Sigma^T U^T)(U\Sigma V^T)$$.

So $$A^TA = V\Sigma^2 V^T$$
